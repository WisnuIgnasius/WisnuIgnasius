# Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Data

```python
df = pd.read_csv('/content/nba2k20-full.csv')
df.head()
```

```python
df.info()
```

Insight = 
1. `team` column has 23 missing values, not necessarily a bad thing, it means those players are currently a free agent.
2. `college` column has 66 missing values, again its not necessarily a bad thing, it means those players didnt go to college / drafted out of highschool. 

## Finding the highest and lowest player's rating.

```python
max_rating = df['rating'].max()
min_rating = df['rating'].min()

print('Highest rating presented in the dataset: ', max_rating)
print('Lowest rating presented in the dataset : ', min_rating)
```

## Top 10 Players by Rating

```python
top10 = df.sort_values(['rating'], ascending=False).head(10)
top10
```

```python
sns.barplot(x = top10['rating'], 
            y = top10['full_name'])
```

## Top 10 lowest players by rating

```python
lowest = df.sort_values(['rating'], ascending=False).tail(10)
lowest
```

```python
sns.barplot(x = lowest['rating'], 
            y = lowest['full_name'])
```

## Average Player's rating

```python
df['rating'].mean()
```

## Players rating distribution

```python
sns.displot(x = df['rating'])
plt.show()
```

## Checking Free Agents by assuming Nan = Free agent

```python
fa = df[df['team'].isna()]
fa
```

```python
fa_num = len(fa.index)
print('The number of Free Agents in this dataset is', fa_num)
```

## Players position distribution 
let's assume that C-F = F-C and F-G = G-F

```python
sns.displot(x = df['position'])
plt.show()
```

## Lets check players nationality

```python
country = df.groupby(['country'],as_index=False).size()
country = country.sort_values(['size'], ascending = False)
country
```

```python
sns.barplot(x = country['country'], 
            y = country['size'])
import matplotlib.pyplot as plt
plt.xticks(rotation=90)
plt.show()
```

## Top 10 country by number of players

```python
country10 = country[:10]
country10
```

```python
sns.barplot(x = country10['country'], 
            y = country10['size'])
plt.xticks(rotation=90)
plt.show()
```

## Top 20 Colleges by number of players

```python
college = df.groupby(['college'],as_index=False).size()
college = college.sort_values(['size'], ascending = False)
college20 = college[:20]
college20
```

```python
sns.barplot(x = college20['college'], 
            y = college20['size'])
plt.xticks(rotation=90)
plt.show()
```

## Players Draft Year distribution

```python
sns.displot(x = df['draft_year'])
```

## Players Draft Round Pie Chart

```python
draftround = df.groupby(['draft_round'],as_index=False).size()
draftround = draftround.sort_values(['size'], ascending = False)
draftround
```

```python
plt.pie(draftround['size'], labels = draftround['draft_round'])
plt.show()
```

## Average rating for every team

```python
team = df.groupby('team')['rating'].mean().reset_index().sort_values('rating', ascending=True)
team
```

```python
sns.barplot(x = team['rating'], 
            y = team['team'])
```

## Average rating for every position

```python
position = df.groupby('position')['rating'].mean().reset_index().sort_values('rating', ascending=True)
position
```

```python
sns.barplot(x = position['rating'], 
            y = position['position'])
```
